**1. User Login Check**
- Problem Statement:
    - Given a username and password, check whether login is successful.
    - username = "admin"
    - password = "1234"
- Requirements:
    - Print "Login Successful" if both username and password match
    - Otherwise print "Invalid Credentials"

Real-World Application: Authentication systems

```python
#Case 1: Correct Credentials
#given credentials
valid_username = "admin"
valid_password = "1234"

#user input
username_input = input("Enter username: ")
password_input = input("Enter password: ")

if username_input==valid_username and password_input==valid_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#Case 2: Wrong Credentials
#given credentials
valid_username = "admin"
valid_password = "1234"

#user input
username_input = input("Enter username: ")
password_input = input("Enter password: ")

if username_input==valid_username and password_input==valid_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

**2. Pass / Fail Analyzer**
- Problem Statement:
    - Analyze student results.
    - marks = [45, 78, 90, 33, 60]
- Requirements:
    - A student passes if marks â‰¥ 50
    - Count the total number of pass students
    - Count the total number of fail students
    - Print the final result clearly

Real-World Application: Academic evaluation systems

```python
marks=list(map(int,input("Enter the marks separated by comma: ").split(",")))
#Given -> marks = [45,78,90,33,60]
total_passed_students = sum(1 for m in marks if m>=50)
total_failed_students = sum(1 for m in marks if m<50)
print(f"Total number of students : {len(marks)}")
print(f"Pass Students : {total_passed_students}")
print(f"Fail Students : {total_failed_students}")
```

**3. Simple Data Cleaner**
- Problem Statement:
    * Clean and standardize user names.
    * names = [" Alice ", "bob", " CHARLIE "]
- Requirements:
    * Remove extra spaces
    * Convert all names to lowercase
    * Print the cleaned list
    
Real-World Application: Data preprocessing before analysis

```python
names=input("Enter names separated by comma: ").split(",")
#Given -> names = [" Alice ", "bob", " CHARLIE "]

cleaned_list=[" ".join(n.strip().split()).title() for n in names]
#split() is used to remove spaces between First and last name of a person | join is used to join first and last name with space in between 
#strip() -> removes spaces in the beginning and end of words
#title() -> converts first letter capital for each word (Standardize user names)
print(cleaned_list)
```

**4. Message Length Analyzer**
- Problem Statement:
    - Analyze message sizes.
    - messages = ["Hi", "Welcome to the platform", "OK"]
- Requirements:
    - Print the length of each message
    - Flag messages longer than 10 characters

Real-World Application: Text filtering and validation systems

```python
messages=input("Enter messages separated by comma: ").split(",")
#Given -> messages = ["Hi", "Welcome to the platform", "OK"]

print(f"{'Messages':<30} | {'Length':<10} | {'Status':<15}")
print("-"*55)

for message in messages:
    length=len(message)
    status="Long" if length>10 else "Short"
    print(f"{message:<30} | {length:<10} | {status:<15}")
```

**5. Error Message Detector**
- Problem Statement:
    - Detect error messages from system logs.
    - logs = ["INFO", "ERROR", "WARNING", "ERROR"]
- Requirements:
    - Count the number of "ERROR" entries
    - Print the total error count

Real-World Application: Monitoring and log analysis systems

```python
logs=input("Enter log entries separated by comma: ").split(",")
#Given -> logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count=sum(1 for log in logs if log.strip().upper()=='ERROR')
print(f"Total error is : {error_count}")
```
